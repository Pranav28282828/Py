```python
d1=[23,56,34,1,65,78,45,34,56,23,57,279,250]
Q11=np.percentile(d1,25)
Q13=np.percentile(d1,75)
print(Q11)
print(Q13)
```

```python
IQR=(Q13-Q11)
print(IQR)
```

```python
LB=Q11-1.5*IQR
UB=Q13+1.5*IQR
print(LB)
print(UB)
```

```python
plt.boxplot(d1)
plt.show()
```

```python
normal_age=[]
for i in d1:
    if i<LB or i>UB:
        print("The Outlier:",i)
    else:
        normal_age.append(i)
print(normal_age)
```
