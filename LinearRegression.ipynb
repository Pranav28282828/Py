```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\prana\Downloads\insurance.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
print(df.isnull().sum().sum())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in df.columns:
    if df[i].dtype=="object":
        df[i]=le.fit_transform(df[i])
```

```python
x=df.drop(columns="charges")
y=df["charges"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size=0.8, random_state=3459)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model_1=LinearRegression()
```

```python
model_1.fit(x_train, y_train)
```

```python
model_answer =model_1.predict(x_test)
```

```python
model_answer
```

```python
from sklearn.metrics import r2_score
print("R2 Score of the model:",r2_score(y_test, model_answer)*100)
```

```python
plt.scatter(y_test, model_answer)
plt.show()
```

```python
sns.regplot(x=y_test, y=model_answer)
plt.show()
```
