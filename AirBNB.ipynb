```python
import matplotlib.pyplot as plt
for i in df.columns:
    if(df[i].dtypes != 'object'):
        plt.boxplot(df[i])
        plt.title(i)
        plt.show()  
```

```python
import numpy as np
def iqr_technique(col):
    Q1=np.percentile(col,25)
    Q3=np.percentile(col,75)
    IQR=Q3-Q1
    LB=Q1-1.5*IQR
    UB=Q3+1.5*IQR
    return LB, UB
```

```python
LB, UB=iqr_technique(df['price'])
df= df[(df['price']>LB) & (df['price']<UB)]
plt.boxplot(df['price'])
plt.title("Price")
plt.show()
```
